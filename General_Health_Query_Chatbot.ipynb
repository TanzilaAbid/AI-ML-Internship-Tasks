```python
import os
import sys
from groq import Groq

# 1. Groq API Key Setup
API_KEY = "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"

try:
    client = Groq(api_key=API_KEY)
except Exception as e:
    print(f"❌ Error initializing Groq client: {e}")

# 2. Define the Chatbot Function
def health_chatbot(user_query):
    system_prompt = """
    You are a friendly, empathetic, and clear AI medical assistant named General Health Query Chatbot.
    Your goal is to provide general, educational information regarding health queries.

    CRITICAL SAFETY RULES:
    1. You are NOT a doctor. Never provide a definitive medical diagnosis or prescribe specific drug dosages.
    2. Always include a brief, polite disclaimer advising the user to consult a healthcare professional for personalized medical advice.
    3. If a user describes life-threatening symptoms (e.g., severe chest pain, difficulty breathing, sudden numbness), immediately urge them to seek emergency medical services.
    4. Keep your language simple, reassuring, and easy to understand for a layperson.
    """

    try:
        # Llama-3.1 model use kar rahe hain jo bilkul active aur updated hai
        completion = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_query}
            ],
            temperature=0.3,
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}"

# 3. Main Chat Loop
def main():
    print("=" * 60)
    print("🏥 General Health Query Chatbot Activated 🏥")
    print("Ask your question or type 'exit' to quit.")
    print("=" * 60)

    while True:
        try:
            user_input = input("\nYou: ").strip()

            if user_input.lower() in ['exit', 'quit']:
                print("General Health Query Chatbot: Take care! Goodbye.")
                break

            if not user_input:
                continue

            print("\nChatbot is thinking...")
            reply = health_chatbot(user_input)
            print(f"\nGeneral Health Query Chatbot:\n{reply}")
            print("-" * 60)

        except KeyboardInterrupt:
            print("\nGeneral Health Query Chatbot: Goodbye!")
            break

if __name__ == "__main__":
    main()
```
